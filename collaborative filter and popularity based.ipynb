```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
matrix = pd.read_csv('matrix.csv',index_col='ProductID')

prod_id = 1
item = matrix.loc[prod_id:prod_id]

def get_similar(item,matrix,original=matrix.copy(),n=10):
    sim = cosine_similarity(item,matrix)
    original['sim'] = sim[0]
    original = original.sort_values(by='sim',ascending=False)
    return original.iloc[1:n+1].index.values
sim = get_similar(item,matrix,matrix.copy(),5)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
matrix = pd.read_csv('matrix.csv',index_col='ProductID')

prod_id = 1
item = matrix.loc[prod_id:prod_id]

def get_similar(item,matrix,original=matrix.copy(),n=10):
    sim = cosine_similarity(item,matrix)
    original['sim'] = sim[0]
    original = original.sort_values(by='sim',ascending=False)
    return original.iloc[1:n+1].index.values
sim = get_similar(item,matrix,matrix.copy(),5)
```

```python
from nltk.corpus import stopwords
import re
```

```python
dfc = pd.read_csv('product.csv')
```

```python
dfc
```

```python
dfc['Description'] = [x.replace("\r\n","") for x in dfc['Description']]
```

```python
dfc
```

```python
list_of_descriptions = []
for product in dfc['Description']:
    product = product.lower()
    desc = re.sub(r'[^a-z ]',' ',product)
    list_of_descriptions.append(desc)
```

```python
list_of_pnames = []
for pname in dfc['ProductName']:
    pname = pname.lower()
    pname = re.sub(r'[^a-z ]',' ',pname)
    list_of_pnames.append(pname)
```

```python
dfc['clean_description'] = list_of_descriptions
dfc['clean_pname'] = list_of_pnames
dfc['text'] = dfc['clean_description'] +' ' + dfc['clean_pname']
```

```python
dfc
```

```python
to_content = dfc[['ProductID','text','Price','quantity','seller_id','hidden']] 
```

```python
to_content
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = CountVectorizer()
```

```python
vectorized_text = vectorizer.fit_transform(to_content['text'])
```

```python
content_df = pd.DataFrame(vectorized_text.toarray(),columns=vectorizer.get_feature_names_out())
```

```python
content_df
```

```python
content_df['price'] = to_content['Price']
content_df['quantity'] = to_content['quantity']
content_df['seller_id'] = to_content['seller_id']
content_df['hidden'] = to_content['hidden']
```

```python
content_df
```

```python
content_df.to_csv('content_matrix.csv',index=False)
```

```python
# deployment code
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
def content_similarity(test,content_df,original,num = 10):
    
    sim = cosine_similarity(test,content_df)
    original['sim'] = sim[0]
    original = original.sort_values(by='sim',ascending=False)
    return original['ProductID'].iloc[1:num]

content_data = pd.read_csv('content_products.csv')
content_matrix = pd.read_csv('content_matrix.csv')

prod_id = 5
prod_details = content_data[content_data['ProductID'] == prod_id]
prod_details_idx = prod_details.index[0]
prod_vector = content_matrix.iloc[prod_details_idx:prod_details_idx+1]
similar_products = content_similarity(prod_vector,content_matrix,original=content_data,num=10)
```
